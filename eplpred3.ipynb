```python
# use loc since you have the label of the row
def geteplteamdata(team_row):

    totalwins = team_row.W
    totaldraws = team_row.D
    totallosses = team_row.L
    totalgoalsscored = team_row.GF
    totalconceeded = team_row.GA
    goalgameratio =team_row.Goal_Game
    defenseratio = team_row.defenseratio
    pointspergame = team_row.PPG
    if totalwins ==0:
        totaldraws=0
        totallosses=0
        totalgoalsscored=0
        totalconceeded=0
        goalgameratio=0
        defenseratio=0
        pointspergame=0
    return[totalwins, totaldraws, totallosses,totalgoalsscored, totalconceeded, goalgameratio, defenseratio, pointspergame]
```

```python
def createteamDict(dataframe):
    teamDict = collections.defaultdict()
    #loop through dataframe rows
    for team in dataframe.itertuples(): 
        teamvector = geteplteamdata(team)
        team_name = team.Teams
        teamDict[team_name] = teamvector
    return teamDict
```
